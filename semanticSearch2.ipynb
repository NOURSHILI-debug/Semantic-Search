```python
from sentence_transformers import SentenceTransformer, util
import numpy as np

documents = [
    "A young boy discovers he is a wizard and attends a magical school.",
    "A team of superheroes must stop an alien invasion.",
    "A man travels back in time to save the future.",
    "A group of friends start a startup in Silicon Valley.",
    "A woman falls in love with a vampire."
]
titles = ["Harry Potter", "Avengers", "Back to the Future", "Silicon Valley", "Twilight"]
```

```python
model=SentenceTransformer('all-MiniLM-L6-v2')
document_embeddings=model.encode(documents, convert_to_tensor=True)
```

```python
query="movie about magic school"
query_embedding=model.encode(query, convert_to_tensor=True)
#compute cosine similarity
cosine_scores = util.cos_sim(query_embedding, document_embeddings)[0]
#Rank results
results =sorted(zip(titles, cosine_scores) ,key=lambda x: x[1], reverse=True)
for title, score, in results:
  print(f"{title}: {score:.4f}")
```
